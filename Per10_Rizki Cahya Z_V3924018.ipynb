```python
# REGRESI LINEAR & BERGANDA (SOAL 1â€“5)
#Rizki Cahya Z_V924018

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# SOAL 1: Jumlah Pelanggan vs Pendapatan

# Dataset: X = jumlah pelanggan, Y = pendapatan (ribu rupiah)
X1 = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
y1 = np.array([200, 400, 600, 800, 1000])
model1 = LinearRegression()
model1.fit(X1, y1)

# Prediksi jika jumlah pelanggan = 35
y1_pred = model1.predict(np.array([[35]]))[0]
print("Soal 1: Y = {:.2f} + {:.2f}X".format(model1.intercept_, model1.coef_[0]))
print(f"Prediksi jika pelanggan 35: {y1_pred:.2f}")
print(f"R^2: {r2_score(y1, model1.predict(X1)):.4f}\n")

# SOAL 2: Waktu Belajar vs Nilai Ujian
# Dataset: X = waktu belajar (jam), Y = nilai ujian
X2 = np.array([1, 2, 3, 4]).reshape(-1, 1)
y2 = np.array([60, 65, 70, 75])
model2 = LinearRegression()
model2.fit(X2, y2)

# Tampilkan persamaan regresi
print("Soal 2: Y = {:.2f} + {:.2f}X".format(model2.intercept_, model2.coef_[0]))
print(f"R^2: {r2_score(y2, model2.predict(X2)):.4f}")
print(f"Rata-rata peningkatan nilai per jam belajar: {model2.coef_[0]:.2f} poin")

# SOAL 3: Usia Kendaraan vs Harga Jual
# Dataset: X = usia (tahun), Y = harga (juta)
X3 = np.array([1, 3, 5, 7, 9]).reshape(-1, 1)
y3 = np.array([150, 120, 90, 60, 30])
model3 = LinearRegression()
model3.fit(X3, y3)

# Prediksi jika usia = 4 tahun
y3_pred = model3.predict(np.array([[4]]))[0]
print("Soal 3: Y = {:.2f} + {:.2f}X".format(model3.intercept_, model3.coef_[0]))
print(f"Prediksi jika usia 4 tahun: {y3_pred:.2f}")
print(f"R^2: {r2_score(y3, model3.predict(X3)):.4f}\n")

# SOAL 4: Jam Internet vs Biaya Pulsa
# Dataset: X = jam, Y = biaya pulsa (rupiah)
X4 = np.array([5, 10, 15, 20]).reshape(-1, 1)
y4 = np.array([25000, 50000, 75000, 100000])
model4 = LinearRegression()
model4.fit(X4, y4)

# Prediksi jika internet 18 jam
y4_pred = model4.predict(np.array([[18]]))[0]
print("Soal 4: Y = {:.2f} + {:.2f}X".format(model4.intercept_, model4.coef_[0]))
print(f"Prediksi jika 18 jam: Rp{y4_pred:.2f}")
print(f"R^2: {r2_score(y4, model4.predict(X4)):.4f}\n")

# SOAL 5: Regresi Linear Berganda
# Dataset: X1 = pengunjung, X2 = menu baru, Y = penjualan
X5 = np.array([
    [50, 2],
    [60, 3],
    [70, 4],
    [80, 5],
    [90, 6]
])
y5 = np.array([400, 480, 560, 640, 720])
model5 = LinearRegression()
model5.fit(X5, y5)

# Prediksi jika pengunjung = 75 dan menu baru = 4
X5_new = np.array([[75, 4]])
y5_pred = model5.predict(X5_new)[0]

print("Soal 5: Y = {:.2f} + {:.2f}*X1 + {:.2f}*X2".format(model5.intercept_, model5.coef_[0], model5.coef_[1]))
print(f"Prediksi jika pengunjung=75, menu=4: {y5_pred:.2f}")
print(f"R^2: {r2_score(y5, model5.predict(X5)):.4f}")
```

```python
# =============================================
# REGRESI LINEAR & BERGANDA DENGAN VISUALISASI
# =============================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from mpl_toolkits.mplot3d import Axes3D

# ===== SOAL 1: Jumlah Pelanggan vs Pendapatan =====
X1 = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
y1 = np.array([200, 400, 600, 800, 1000])
model1 = LinearRegression().fit(X1, y1)

plt.figure()
plt.scatter(X1, y1, color='blue', label='Data Asli')
plt.plot(X1, model1.predict(X1), color='red', label='Regresi')
plt.title("Soal 1: Pelanggan vs Pendapatan")
plt.xlabel("Jumlah Pelanggan")
plt.ylabel("Pendapatan (ribu)")
plt.legend()
plt.grid(True)
plt.show()

# ===== SOAL 2: Waktu Belajar vs Nilai Ujian =====
X2 = np.array([1, 2, 3, 4]).reshape(-1, 1)
y2 = np.array([60, 65, 70, 75])
model2 = LinearRegression().fit(X2, y2)

plt.figure()
plt.scatter(X2, y2, color='blue', label='Data Asli')
plt.plot(X2, model2.predict(X2), color='red', label='Regresi')
plt.title("Soal 2: Waktu Belajar vs Nilai")
plt.xlabel("Jam Belajar")
plt.ylabel("Nilai Ujian")
plt.legend()
plt.grid(True)
plt.show()

# ===== SOAL 3: Usia Kendaraan vs Harga Jual =====
X3 = np.array([1, 3, 5, 7, 9]).reshape(-1, 1)
y3 = np.array([150, 120, 90, 60, 30])
model3 = LinearRegression().fit(X3, y3)

plt.figure()
plt.scatter(X3, y3, color='blue', label='Data Asli')
plt.plot(X3, model3.predict(X3), color='red', label='Regresi')
plt.title("Soal 3: Usia Kendaraan vs Harga Jual")
plt.xlabel("Usia (tahun)")
plt.ylabel("Harga (juta)")
plt.legend()
plt.grid(True)
plt.show()

# ===== SOAL 4: Jam Internet vs Biaya Pulsa =====
X4 = np.array([5, 10, 15, 20]).reshape(-1, 1)
y4 = np.array([25000, 50000, 75000, 100000])
model4 = LinearRegression().fit(X4, y4)

plt.figure()
plt.scatter(X4, y4, color='blue', label='Data Asli')
plt.plot(X4, model4.predict(X4), color='red', label='Regresi')
plt.title("Soal 4: Jam Internet vs Biaya Pulsa")
plt.xlabel("Jam Internet")
plt.ylabel("Biaya (rupiah)")
plt.legend()
plt.grid(True)
plt.show()

# ===== SOAL 5: Regresi Linear Berganda (3D) =====
X5 = np.array([
    [50, 2],
    [60, 3],
    [70, 4],
    [80, 5],
    [90, 6]
])
y5 = np.array([400, 480, 560, 640, 720])
model5 = LinearRegression().fit(X5, y5)

# Buat permukaan prediksi
x1_range = np.linspace(50, 90, 10)
x2_range = np.linspace(2, 6, 10)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)
x_pred = np.c_[x1_grid.ravel(), x2_grid.ravel()]
y_pred = model5.predict(x_pred).reshape(x1_grid.shape)

# Plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X5[:, 0], X5[:, 1], y5, color='blue', label='Data Asli')
ax.plot_surface(x1_grid, x2_grid, y_pred, cmap='viridis', alpha=0.5)
ax.set_title("Soal 5: Penjualan vs Pengunjung & Menu Baru")
ax.set_xlabel("Pengunjung (X1)")
ax.set_ylabel("Menu Baru (X2)")
ax.set_zlabel("Penjualan (Y)")
plt.legend()
plt.show()
```
