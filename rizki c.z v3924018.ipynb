```python
import matplotlib.pyplot as plt
import numpy as np

# Soal No.1: Linear Regression
x1 = np.array([10, 20, 30, 40, 50])
y1 = np.array([200, 400, 600, 800, 1000])
m1, b1 = 20, 0  # y = 20x

# Soal No.2: Linear Regression
x2 = np.array([1, 2, 3, 4])
y2 = np.array([60, 65, 70, 74])
m2, b2 = 5, 55  # y = 5x + 55

# Soal No.3: Linear Regression
x3 = np.array([1, 3, 5, 7, 9])
y3 = np.array([150, 120, 90, 60, 30])
m3, b3 = -15, 165  # y = -15x + 165

# Soal No.4: Linear Regression
x4 = np.array([5, 10, 15, 20])
y4 = np.array([25000, 50000, 75000, 100000])
m4, b4 = 5000, 0  # y = 5000x

# Soal No.5: Multiple Linear Regression Visualization (Basic 2D)
x5_1 = np.array([50, 60, 70, 80, 90])   # Feature 1: Waktu Layanan
x5_2 = np.array([2, 3, 4, 5, 6])        # Feature 2: Jumlah Komplain
y5 = np.array([400, 480, 560, 640, 720])# Target: Tingkat Kepuasan
a5, b5_1, b5_2 = -320, 8, 80  # Regression equation: Y = -320 + 8X1 + 80X2

# Create 2x2 subplot for first four problems
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot Soal 1: Simple linear regression
axs[0, 0].scatter(x1, y1, color='blue')
axs[0, 0].plot(x1, m1 * x1 + b1, color='red')
axs[0, 0].set_title("No.1 : Y = 20X")
axs[0, 0].set_xlabel("Pelanggan")
axs[0, 0].set_ylabel("Pendapatan (ribu)")

# Plot Soal 2: Simple linear regression with intercept
axs[0, 1].scatter(x2, y2, color='green')
axs[0, 1].plot(x2, m2 * x2 + b2, color='red')
axs[0, 1].set_title("No.2 : Y = 5X + 55")
axs[0, 1].set_xlabel("Jam Belajar")
axs[0, 1].set_ylabel("Nilai")

# Plot Soal 3: Negative slope linear regression
axs[1, 0].scatter(x3, y3, color='purple')
axs[1, 0].plot(x3, m3 * x3 + b3, color='red')
axs[1, 0].set_title("No.3 : Y = -15X + 165")
axs[1, 0].set_xlabel("Umur Kendaraan (tahun)")
axs[1, 0].set_ylabel("Nilai Kendaraan (juta)")

# Plot Soal 4: Large scale linear regression
axs[1, 1].scatter(x4, y4, color='orange')
axs[1, 1].plot(x4, m4 * x4 + b4, color='red')
axs[1, 1].set_title("No.4 : Y = 5000X")
axs[1, 1].set_xlabel("Jam Kerja")
axs[1, 1].set_ylabel("Pendapatan (rupiah)")

# Show all subplots nicely spaced
plt.tight_layout()

# Basic 2D visualization for Soal 5:
# Plot x5_1 (Waktu Layanan) vs y5 (Tingkat Kepuasan)
# Use color to represent x5_2 (Jumlah Komplain) - the second feature
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x5_1, y5, c=x5_2, cmap='winter', s=100, edgecolor='black')
plt.colorbar(scatter, label='X2: Jumlah Komplain')
plt.title("No.5 : Visualisasi Regresi Linier Berganda (2D Basic)")
plt.xlabel("X1: Waktu Layanan")
plt.ylabel("Tingkat Kepuasan")
plt.grid(True)

# Explanation text printed to console for user understanding
explanation = """
Penjelasan Visualisasi No.5:
- Ini adalah visualisasi regresi linier berganda dengan dua variabel bebas (X1 dan X2).
- Sumbu X1 (Waktu Layanan) diplot pada sumbu horizontal.
- Sumbu Y (Tingkat Kepuasan) diplot pada sumbu vertikal.
- Warna titik mewakili nilai X2 (Jumlah Komplain) menggunakan skala warna.
- Dengan cara ini, kita dapat melihat bagaimana Tingkat Kepuasan dipengaruhi oleh kombinasi Waktu Layanan dan Jumlah Komplain.
- Titik dengan warna berbeda menunjukkan variasi variabel kedua (X2).
"""

print(explanation)

plt.show()
```
